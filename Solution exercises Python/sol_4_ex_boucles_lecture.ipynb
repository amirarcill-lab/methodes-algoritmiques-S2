#### Exercice 1 - Transformation d'une liste

On considère la liste suivante :

```python
list_of_notes=[18, 12, 13, 0, 6, 19, 13, 5, 12, 19, 8, 19, 8, 16, 18, 16, 11, 7, 16, 12, 19, 15, 12, 1, 10, 19, 16, 9, 1, 3]
```

Crée une nouvelle liste notes_majorees qui contient chaque note augmentée de 1 point.

```python
list_of_notes_maj=[i+1 for i in list_of_notes]
```

#### Exercice 2 - Filtrer une liste

À partir de la liste_notes_majorees, crée une liste notes_valides qui contient uniquement les notes supérieures ou égales à 10

```python
notes_valides=[i for i in list_of_notes_maj if i>=10]
```

Combien d'étudiants ont et une note supérieure à 10?

```python
len(notes_valides)
```

#### Exercice 3 - Filtrer une liste 2

À partir de la liste_notes_majorees, crée une liste notes_sup_moy qui contient uniquement les notes supérieures à la moyenne de la classe.

```python
import numpy as np
moy=np.mean(notes_valides)
notes_sup_moy=[i for i in notes_valides if i>moy]
print(notes_sup_moy)
```

#### Exercice 7 – Lecture de texte et transformation

Créer une liste de textes en lisant le fichier sim_tweets.txt ligne par ligne.

Préparer chaque élément de la liste pour une analyse de texte en supprimant la ponctuation et en mettant tout en minuscules.

```python
# fonction de nettoyage de texte (defini dans le cours 5)

import string
punctuation_chars = list(string.punctuation)

def nettoyer_txt(txt,punct):
    for p in punct:
        txt=txt.replace(p,'') #remplace chaque caractère de ponctuation par une chaîne vide
    txt=txt.lower() #met tout en minuscules

    return txt #return le texte nettoyé
```

```python
infile='../data/sim_tweets.txt' # attention,le chemnin peut varier selon votre organisation des dossiers
data=[]

for line in open(infile):
    data.append(line.strip())
```

```python
data[0]
```

```python
nettoyer_txt(data[0], punctuation_chars) #exemople d'utilisation de la fonction sur un element de la liste data
```

```python
#application de la fonction à toute la liste data
data_clean=[nettoyer_txt(text,punctuation_chars) for text in data]
```
