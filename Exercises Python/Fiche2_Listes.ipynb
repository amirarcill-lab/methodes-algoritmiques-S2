## Exercice 1 – Création de listes

### Subtask:
Créez une liste nommée 'prenoms' avec cinq prénoms de votre choix et une liste nommée 'ages' avec les âges correspondants.


```python
prenoms = ["Alice", "Bob", "Charlie", "Diana", "Eve"]
ages = [25, 30, 22, 28, 35]

print("Liste des prénoms:", prenoms)
print("Liste des âges:", ages)
```

## Exercice 2 – Accès aux éléments

### Subtask:
Affichez le premier élément de la liste 'prenoms' et le dernier élément de la liste 'ages'.


```python
print("Premier prénom:", prenoms[0])
print("Dernier âge:", ages[-1])
```

## Exercice 3 – Longueur d’une liste

### Subtask:
Calculez et affichez le nombre d’éléments contenus dans la liste 'prenoms'.


```python
nombre_prenoms = len(prenoms)
print(len(prenoms))
```

## Exercice 4 – Ajout d’éléments

### Subtask:
Ajoutez un nouveau prénom à la fin de la liste 'prenoms' et affichez la liste mise à jour.


```python
prenoms.append("Frank")
print("Liste des prénoms mise à jour:", prenoms)
```

## Exercice 5 – Modification d’un élément

### Subtask:
Remplacez le deuxième prénom de la liste 'prenoms' par un autre prénom de votre choix et affichez la liste.


```python
prenoms[1] = "Bernard"
print(prenoms)
```

## Exercice 6 – Listes et chaînes de caractères

### Subtask:
Transformez la liste `prenoms` en une seule chaîne de caractères, toute en minuscule, dans laquelle les prénoms sont séparés par un tiret (ex. : lucille-amadou-sarah-karim).

```python
prenoms_minuscules = [prenom.lower() for prenom in prenoms]
chaine_prenoms = "-".join(prenoms_minuscules)
print(chaine_prenoms)
```

## Exercice 7 – Vérification de présence

### Subtask:
Vérifiez si le prénom "Lucille" est présent dans la liste prenoms.

```python
if "Lucille" in prenoms:
    print("oui")
else:
    print("non")
```

## Exercice 8 – Composition de listes

### Subtask:
- Créez une deuxième liste appelée `prenoms2`.
- Créez une liste `tousPrenoms` qui regroupe les deux listes `prenoms` et `prenoms2`.
- Créez une liste de listes `listesDePrenoms` qui contient comme éléments les listes `prenoms` et `prenoms2`.

```python
prenoms2 = ["Georges", "Hannah", "Isabelle"]
tousPrenoms = prenoms + prenoms2
listesDePrenoms = [prenoms, prenoms2]

print(prenoms2)
print(tousPrenoms)
print(listesDePrenoms)
```

## Exercice 9 - liste des listes
- Affichez la première liste contenue dans listesDePrenoms.
- Affichez ensuite le deuxieme prénom de cette liste.

```python
premiere_liste = listesDePrenoms[0]
print(premiere_liste)

deuxieme_prenom_premiere_liste = premiere_liste[1]
print(deuxieme_prenom_premiere_liste)
```
