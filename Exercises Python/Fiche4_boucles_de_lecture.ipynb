Exercice 1 - Transformation d'une liste
On considère la liste suivante :

`list_of_notes=[18, 12, 13, 0, 6, 19, 13, 5, 12, 19, 8, 19, 8, 16, 18, 16, 11, 7, 16, 12, 19, 15, 12, 1, 10, 19, 16, 9, 1, 3]`
Crée une nouvelle liste `notes_majorees` qui contient chaque note augmentée de 1 point.

````python
list_of_notes=[18, 12, 13, 0, 6, 19, 13, 5, 12, 19, 8, 19, 8, 16, 18, 16, 11, 7, 16, 12, 19, 15, 12, 1, 10, 19, 16, 9, 1, 3]
list_of_notes_maj=[i+1 for i in list_of_notes]
print(list_of_notes_maj)
````

Exercice 2 - Filtrer une liste
À partir de la `liste_notes_majorees`, crée une liste `notes_valides` qui contient uniquement les notes supérieures ou égales à 10.
Combien d'étudiants ont une note supérieure ou égale à 10?

````python
notes_valides=[i for i in list_of_notes_maj if i>=10]
print(notes_valides)
print(f"Nombre d'étudiants avec une note supérieure ou égale à 10: {len(notes_valides)}")
````

Exercice 3 - Filtrer une liste 2
À partir de la `liste_notes_majorees`, crée une liste `notes_sup_moy` qui contient uniquement les notes supérieures à la moyenne de la classe.

````python
import numpy as np
moy=np.mean(notes_valides)
notes_sup_moy=[i for i in notes_valides if i>moy]
print(notes_sup_moy)
````

Exercice 4 - Listes de listes
On considère la liste suivante :

```python
personnes = [
    ["Marie", "Durand"],
    ["Lucas", "Martin"],
    ["Anna", "Petit"],
    ["Sarah", "Leroy"]
]
```
a- Extraire la liste des prénoms.
b- Extraire la liste des noms de famille.

````python
personnes = [
    ["Marie", "Durand"],
    ["Lucas", "Martin"],
    ["Anna", "Petit"],
    ["Sarah", "Leroy"]
]

prenoms=[i[0] for i in personnes]
print(prenoms)

noms=[i[1] for i in personnes]
print(noms)
````
